```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from data_engine import DataEngine
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

C = {
    'bg': '#F4F4F5',
    'card': '#FFFFFF',
    'card_border': 'rgba(15, 23, 42, 0.05)',
    'card_shadow': '0 10px 30px rgba(0, 0, 0, 0.02)',
    'card_radius': '24px',
    'primary': '#1E40AF',
    'accent': '#4F46E5',
    'text': '#1E293B',
    'text_dark': '#0F172A',
    'muted': '#64748B',
    'label': '#94A3B8',
    'divider': 'rgba(15, 23, 42, 0.05)',
    'bar_bg': 'rgba(15, 23, 42, 0.04)',
    'icon_bg': 'rgba(100, 116, 139, 0.063)',
    'success': '#10B981',
    'danger': '#EF4444',
    'warning': '#F59E0B'
}
FD = "'Outfit', sans-serif"
FONT = "'Inter', sans-serif"

_base = dict(
    template='plotly_white',
    plot_bgcolor=C['card'],
    paper_bgcolor=C['card'],
    margin=dict(l=50, r=30, t=80, b=50),
    font=dict(family='Inter, sans-serif', color=C['muted'], size=12)
)

def _style(fig, title, sub='', h=400):
    t = f'<b style="font-family:{FD};font-size:16px;font-weight:700;color:{C["text"]};letter-spacing:-0.01em;">{title}</b>'
    if sub:
        t += f'<br><span style="font-family:{FONT};font-size:12px;font-weight:500;color:{C["muted"]};">{sub}</span>'
    fig.update_layout(**_base, height=h, title=dict(text=t, x=0.02, y=0.95))
    fig.update_xaxes(gridcolor=C['divider'], zeroline=False, title_font=dict(size=12, color=C['muted']))
    fig.update_yaxes(gridcolor=C['divider'], zeroline=False, title_font=dict(size=12, color=C['muted']))
    return fig

def section_title(title, subtitle):
    display(HTML(f'''
        <div style="margin:50px 0 24px; background:{C['card']}; padding:20px 24px; border-radius:{C['card_radius']}; border:1px solid {C['card_border']}; box-shadow:{C['card_shadow']}; font-family:{FONT};">
            <h2 style="margin:0; color:{C['text']}; font-size:20px; font-family:{FD}; font-weight:700; letter-spacing:-0.02em;">{title}</h2>
            <p style="margin:4px 0 0; color:{C['muted']}; font-size:12px; font-weight:500;">{subtitle}</p>
        </div>
    '''))

def render_kpis(kpis):
    html = f'<div style="display:grid; grid-template-columns: repeat(auto-fit, minmax(220px, 1fr)); gap:24px; margin-bottom:32px; font-family:{FONT};">'
    for label, val, icon, accent in kpis:
        html += f'''
            <div style="background:{C['card']}; padding:24px; border-radius:{C['card_radius']}; border:1px solid {C['card_border']}; box-shadow:{C['card_shadow']};">
                <div style="display:flex; align-items:center; gap:12px; margin-bottom:20px;">
                    <div style="width:40px; height:40px; border-radius:12px; background:{accent}10; display:flex; align-items:center; justify-content:center; font-size:20px;">
                        {icon}
                    </div>
                    <span style="font-size:11px; font-weight:700; color:{C['label']}; text-transform:uppercase; letter-spacing:0.05em;">{label}</span>
                </div>
                <div style="font-family:{FD}; font-size:32px; font-weight:800; color:{C['text']}; letter-spacing:-0.04em; line-height:1; font-feature-settings:'tnum' 1,'lnum' 1;">{val}</div>
            </div>
        '''
    html += '</div>'
    display(HTML(html))

def render_podium(items, title, accent=C['accent'], suffix=""):
    html = f'<div style="flex:1; background:{C["card"]}; border:1px solid {C["card_border"]}; border-radius:{C["card_radius"]}; padding:24px; font-family:{FONT}; box-shadow:{C["card_shadow"]}; min-width:280px;">'
    html += f'<h3 style="margin:0 0 20px 0; color:{C["text"]}; font-family:{FD}; font-size:16px; font-weight:700; letter-spacing:-0.02em;">{title}</h3>'
    html += '<div style="display:flex; flex-direction:column; gap:14px;">'
    for i, (name, val) in enumerate(items):
        max_val = max(v for _, v in items) if items else 1
        pct = int((val / max_val) * 100) if max_val > 0 else 0
        html += f'''
            <div style="display:flex; flex-direction:column; gap:6px;">
                <div style="display:flex; justify-content:space-between; align-items:center;">
                    <span style="font-size:13px; font-weight:600; color:{C['text']};">{name}</span>
                    <span style="font-size:12px; font-weight:700; color:{C['muted']}; font-feature-settings:'tnum' 1;">{val} {suffix}</span>
                </div>
                <div style="width:100%; height:6px; background:{C['bar_bg']}; border-radius:99px; overflow:hidden;">
                    <div style="height:100%; width:{pct}%; background:{accent}; border-radius:99px;"></div>
                </div>
            </div>
        '''
    html += '</div></div>'
    return html

def render_insight(title, text):
    display(HTML(f'''
        <div style="background:{C['card']}; border:1px solid {C['card_border']}; border-left:3px solid {C['accent']}; padding:20px 24px; border-radius:0 {C['card_radius']} {C['card_radius']} 0; font-family:{FONT}; margin:24px 0 32px; box-shadow:{C['card_shadow']};">
            <div style="display:flex; align-items:center; gap:8px; margin-bottom:8px;">
                <span style="font-size:14px;">💡</span>
                <span style="font-size:11px; font-weight:700; color:{C['label']}; text-transform:uppercase; letter-spacing:0.05em;">{title}</span>
            </div>
            <div style="color:{C['text']}; font-size:13px; line-height:1.6; font-weight:400;">{text}</div>
        </div>
    '''))

def render_alert(title, text, type='warning'):
    color = C['danger'] if type == 'danger' else C['success'] if type == 'success' else C['warning']
    icon = '⚠️' if type == 'danger' else '✅'
    display(HTML(f'''
        <div style="background:{C['card']}; border:1px solid {C['card_border']}; border-left:3px solid {color}; padding:20px 24px; border-radius:0 {C['card_radius']} {C['card_radius']} 0; font-family:{FONT}; margin:24px 0; box-shadow:{C['card_shadow']};">
            <div style="display:flex; align-items:center; gap:8px; margin-bottom:8px;">
                <span style="font-size:14px;">{icon}</span>
                <span style="font-size:11px; font-weight:700; color:{color}; text-transform:uppercase; letter-spacing:0.05em;">{title}</span>
            </div>
            <div style="color:{C['text']}; font-size:13px; line-height:1.6; font-weight:400;">{text}</div>
        </div>
    '''))

print("✅ Entorno inicializado correctamente")
```

```python
engine = DataEngine()
df_emp = engine.get_employees()
df_prj = engine.get_projects()
df_asg = engine.get_assignments()

df_emp['Antigüedad'] = df_emp['ANTIGUEDAD_ANOS'].fillna(0).astype(int)
df_emp['Edad'] = df_emp['EDAD'].fillna(0).astype(int)
df_emp['AnioAlta'] = pd.to_datetime(df_emp['F_ALTA']).dt.year
df_emp['AnioBaja'] = pd.to_datetime(df_emp['F_BAJA']).dt.year

activos = df_emp[df_emp['ESTADO'] == 'Activo'].copy()
prj_act = df_prj[df_prj['ESTADO'] == 'Activo'].copy()

edad_med = int(round(activos['Edad'].mean()))
prj_fin = len(df_prj[df_prj['ESTADO'] != 'Activo'])

render_kpis([
    ('Plantilla Activa', len(activos), '👥', '#64748B'),
    ('Proyectos Activos', len(prj_act), '🚀', '#64748B'),
    ('Proyectos Finalizados', prj_fin, '🏁', '#64748B'),
    ('Edad Promedio', f"{edad_med} años", '🎂', '#64748B')
])
```

```python
section_title('Equipo', 'Antigüedad y nuevas altas')

top_veteranos = activos.nlargest(5, 'Antigüedad')[['TX_NOMBRE', 'Antigüedad']].values.tolist()
top_recientes = activos.nsmallest(5, 'Antigüedad')[['TX_NOMBRE', 'Antigüedad']].values.tolist()

podios_html = f'<div style="display:flex; gap:24px; flex-wrap:wrap; margin-bottom:32px;">'
podios_html += render_podium(top_veteranos, 'Mayor antigüedad', C['accent'], 'años')
podios_html += render_podium(top_recientes, 'Últimas altas', C['success'], 'años')
podios_html += '</div>'
display(HTML(podios_html))

render_insight(
    'Experiencia en el equipo',
    f'El análisis muestra una base de experiencia sólida con personas que llevan hasta <b>{int(top_veteranos[0][1])} años</b>. Simultáneamente, las {len(top_recientes)} altas recientes permiten una renovación constante del equipo.'
)

fig_ant = px.histogram(activos, x='Antigüedad', nbins=15, 
                       color_discrete_sequence=[C['accent']], opacity=0.85)
fig_ant.update_traces(marker_line_width=0, hovertemplate='Años: %{x}<br>Empleados: %{y}<extra></extra>')
_style(fig_ant, 'Años en la empresa', h=380)
fig_ant.update_yaxes(title_text="Empleados")
fig_ant.update_xaxes(title_text="Años")
fig_ant.update_layout(bargap=0.15)
fig_ant.show()
```

```python
section_title('Demografía', 'Edad y estado civil')

df_civil = activos[activos['CX_EDOCIVIL'].isin(['S', 'C'])].copy()
df_civil['Estado_Civil_Full'] = df_civil['CX_EDOCIVIL'].map({'S': 'Soltero/a', 'C': 'Casado/a'})

fig_civil = px.pie(df_civil, names='Estado_Civil_Full', hole=0.6, 
                   color_discrete_sequence=[C['accent'], C['success']])
fig_civil.update_traces(textinfo='percent', textfont_size=13, 
                        marker=dict(line=dict(color=C['card'], width=2)),
                        hovertemplate='<b>%{label}</b><br>%{percent}<extra></extra>')
_style(fig_civil, 'Estado civil', h=380)
fig_civil.update_layout(showlegend=True, legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1, font=dict(size=12)))
fig_civil.show()

fig_edad = px.histogram(activos, x='Edad', nbins=12, marginal="box", 
                        color_discrete_sequence=[C['warning']], opacity=0.85)
fig_edad.update_traces(marker_line_width=0, hovertemplate='Edad: %{x}<br>Empleados: %{y}<extra></extra>')
_style(fig_edad, 'Distribución de edad', h=380)
fig_edad.update_yaxes(title_text="Empleados")
fig_edad.update_xaxes(title_text="Edad")
fig_edad.update_layout(bargap=0.15)
fig_edad.show()

render_insight(
    'Datos demográficos',
    f'La media de edad se sitúa en <b>{edad_med} años</b>. Es un equipo joven, algo que también se refleja en la alta proporción de personas solteras, un patrón común en estas edades.'
)
```

```python
section_title('Histórico', 'Evolución de contrataciones')

altas = df_emp.groupby('AnioAlta').size().reset_index(name='Altas')
bajas = df_emp.dropna(subset=['AnioBaja']).groupby('AnioBaja').size().reset_index(name='Bajas')
evol = pd.merge(altas, bajas, left_on='AnioAlta', right_on='AnioBaja', how='outer').fillna(0)
evol['Año'] = evol['AnioAlta'].combine_first(evol['AnioBaja']).astype(int)
evol = evol.sort_values('Año')
evol['Neto'] = evol['Altas'] - evol['Bajas']

best_y = int(evol.loc[evol['Neto'].idxmax()]['Año'])
worst_y = int(evol.loc[evol['Neto'].idxmin()]['Año'])

fig_evol = go.Figure()
fig_evol.add_trace(go.Scatter(x=evol['Año'], y=evol['Altas'], name='Altas', 
                              mode='lines+markers', fill='tozeroy', 
                              line=dict(color=C['accent'], width=2.5),
                              fillcolor='rgba(79, 70, 229, 0.08)',
                              marker=dict(size=7, color=C['card'], line=dict(color=C['accent'], width=2))))
fig_evol.add_trace(go.Scatter(x=evol['Año'], y=evol['Bajas'], name='Bajas', 
                              mode='lines+markers', fill='tozeroy', 
                              line=dict(color=C['danger'], width=2.5),
                              fillcolor='rgba(239, 68, 68, 0.06)',
                              marker=dict(size=7, color=C['card'], line=dict(color=C['danger'], width=2))))

_style(fig_evol, 'Altas frente a bajas por año', h=380)
fig_evol.update_xaxes(dtick=1, title_text="") 
fig_evol.update_yaxes(title_text="Empleados")
fig_evol.update_layout(hovermode="x unified", legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1, font=dict(size=12)))
fig_evol.show()

render_insight(
    'Tendencias de plantilla',
    f'Revisando el histórico, el año <b>{best_y}</b> destacó por su fuerte crecimiento de equipo. Por contra, el año <b>{worst_y}</b> registró un pico de salidas que convendría tener en cuenta de cara a afinar la retención a futuro.'
)
```

```python
section_title('Proyectos', 'Duración y sedes')

df_prj['Duracion_Int'] = df_prj['DURACION_DIAS'].fillna(0).astype(int)

fig_dur = px.histogram(df_prj, x='Duracion_Int', nbins=15, 
                       color_discrete_sequence=[C['accent']], opacity=0.85)
fig_dur.update_traces(marker_line_width=0, hovertemplate='Días: %{x}<br>Proyectos: %{y}<extra></extra>')
_style(fig_dur, 'Duración planificada de proyectos', h=360)
fig_dur.update_yaxes(title_text="Proyectos")
fig_dur.update_xaxes(title_text="Días")
fig_dur.update_layout(bargap=0.15)
fig_dur.show()

sedes = df_prj.groupby('TX_LUGAR').size().reset_index(name='Total').sort_values('Total', ascending=False)
top_sedes = sedes.head(5)[['TX_LUGAR', 'Total']].values.tolist()

html_podio = f'<div style="display:flex; margin-bottom: 24px;">' + render_podium(top_sedes, 'Top sedes por volumen operativo', C['accent'], 'proyectos') + '</div>'
display(HTML(html_podio))

render_insight(
    'Concentración geográfica',
    f'Las cifras indican que la mayor concentración operativa se encuentra en <b>{str(sedes.iloc[0]["TX_LUGAR"])}</b>. Saber esto resulta muy útil para planificar mejor la asignación de recursos y el presupuesto de los próximos trimestres.'
)
```

```python
section_title('Asignaciones', 'Carga de trabajo y avisos')

asg_activos = df_asg[df_asg['ID_EMPLEADO'].isin(activos['ID_EMPLEADO'])]
emp_projs = asg_activos.groupby('ID_EMPLEADO').size().reset_index(name='Num_Proyectos')

dist_carga = emp_projs.groupby('Num_Proyectos').size().reset_index(name='Personas')
dist_carga['Etiqueta'] = dist_carga['Num_Proyectos'].apply(lambda x: f"Con {int(x)} proyecto{'s' if x>1 else ''}")

fig_carga = px.bar(dist_carga, x='Personas', y='Etiqueta', orientation='h',
                   color_discrete_sequence=[C['accent']])
_style(fig_carga, 'Nivel de carga por persona', h=280)
fig_carga.update_traces(text=dist_carga['Personas'], textposition='outside', textfont=dict(size=12, color=C['muted']), 
                        hovertemplate='%{y}<br>Personas: %{x}<extra></extra>', opacity=0.75, marker_line_width=0)
fig_carga.update_xaxes(showgrid=False, zeroline=False, showticklabels=False, title_text="")
fig_carga.update_yaxes(title_text="")
fig_carga.update_layout(margin=dict(l=200, r=60, t=70, b=30))
fig_carga.show()

proj_counts = df_asg.groupby('ID_PROYECTO').size().reset_index(name='Empleados')
proj_counts = pd.merge(proj_counts, df_prj[['ID_PROYECTO', 'TX_DESCRIPCION']], on='ID_PROYECTO')
top_proyectos = proj_counts.nlargest(5, 'Empleados')[['TX_DESCRIPCION', 'Empleados']].values.tolist()

display(HTML(render_podium(top_proyectos, 'Proyectos que requieren más personal', C['warning'], 'personas')))

render_insight(
    'Carga del equipo',
    f'En general, las asignaciones están equilibradas y la gente participa en 1 o 2 iniciativas. No obstante, se nota que proyectos como <b>{top_proyectos[0][0]}</b> demandan gran parte del equipo y son críticos operativamente.'
)

asignados_ids = df_asg['ID_PROYECTO'].unique()
proyectos_huerfanos = df_prj[(~df_prj['ID_PROYECTO'].isin(asignados_ids)) & (df_prj['ESTADO'] == 'Activo')]

if len(proyectos_huerfanos) > 0:
    nombres = " · ".join(proyectos_huerfanos['TX_DESCRIPCION'].tolist())
    render_alert('Revisión Pendiente', 
                 f'Durante el cruce de datos, se han detectado <b>{len(proyectos_huerfanos)} proyecto(s)</b> en estado activo pero sin nadie asignado. Sería conveniente revisar su estado:<br><br><b>{nombres}</b>', type='danger')
else:
    render_alert('Asignaciones al 100%', 
                 f'No se detectan proyectos activos sin personal. Todas las asignaciones están al día y cruzadas correctamente.', type='success')
```
